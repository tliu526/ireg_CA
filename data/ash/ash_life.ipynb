```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import pandas as pd
```

#Ash Life Analysis

##Small Grid Size (25 by 25)

```python
life_s = pd.read_csv('life_test_s.csv', sep=' ')

#print life_s
avg_life = []

for i in range(1,100):
    test = life_s[life_s['Initial_Percent'] == i]
    #print len(test)
    avg_life.append(float(sum(test['Time']))/float(len(test)))
    
y = np.array(avg_life)
x = np.arange(1,100,1)

plt.plot(x,y)
x1,x2,y1,y2 = plt.axis()
plt.title("Lifetime to Stability against Initial Density (Small Voronoi)")
plt.xlabel('Initial Percentage of ON States')
plt.ylabel('Mean Lifetime to Stability')
plt.savefig('Lifetime_Stability_S.jpg')
```

```python
density1 = life_s[life_s['Initial_Percent'] == 20]
x1 = np.array(density1['Percent_On'])
y1 = np.array(density1['Time'])

density2 = life_s[life_s['Initial_Percent'] == 40]
x2 = np.array(density2['Percent_On'])
y2 = np.array(density2['Time'])

density3 = life_s[life_s['Initial_Percent'] == 60]
x3 = np.array(density3['Percent_On'])
y3 = np.array(density3['Time'])

plt.scatter(x1,y1,c='r', marker='s', label='20%')
plt.scatter(x2,y2,c='b', marker='o', label='40%')
plt.scatter(x3,y3,c='y', marker='^', label='60%')

plt.title("Ash Density against Lifetime (Small Voronoi)")
plt.xlabel('Ash Density')
plt.ylabel('Lifetime')
plt.legend(loc='upper left')
plt.savefig('Ash_Density_S.jpg')
```
