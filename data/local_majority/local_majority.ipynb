```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import pandas as pd
```

#Local Majority Analysis

##Regular Grids (Non-periodic)

```python
lm_reg = pd.read_csv('LM_Reg.csv', sep=' ')

print lm_reg.head()
correct_counts = []

for i in range(1, 100):
    test = lm_reg[lm_reg['Initial_Percent'] == i]
    correct_counts.append(float(sum(test['Correct_Class'])) / float(100))
    
#print correct_counts
y = np.array(correct_counts)
x = np.arange(1,100,1)

plt.plot(x,y)
x1,x2,y1,y2 = plt.axis()
plt.axis((x1,x2,0,1.1))
plt.title("Baseline Non-Periodic Regular LM Performance (100 runs)")
plt.xlabel('Initial Percentage of ON States')
plt.ylabel('Fraction Task-Performance')
plt.savefig('LM_Baseline_Reg.jpg')
```
